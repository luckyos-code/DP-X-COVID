```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Let's create a Dataframe using lists
budgets = ['0.1', '1', '10', 'inf']

resnet18 = [69.6, 74.1, 69.3, 77.5]
resnet50_img = [0, 0, 0, 97.6]
resnet50_pneu = [72.0, 73.5, 71.6, 77.9]
shallow = [0, 0, 0, 79.6]
PATE = [40.0, 40.7, 74.9, 79.9]

# Now, create a pandas dataframe using above lists
df_acc = pd.DataFrame(
    {'Privacy Budget' : budgets,
     'ResNet50-ImageNet' : resnet50_img,
     'ResNet18-tanh-Pneumonia' : resnet18,
     'ResNet50-tanh-Pneumonia' : resnet50_pneu,
     'ShallowNetwork' : shallow,
     'PATE-25-ResNet50-Pneumonia' : PATE,
     })

df_acc
```

```python
#Plot a line graph
fig, ax = plt.subplots(figsize=[10, 8])

for x in df_acc:
    if x != 'Privacy Budget':
        ax.plot(df_acc['Privacy Budget'],
            df_acc[x],
            marker='x', linewidth=2, label=x)

plt.ylim([0, 100])

# Add labels and title
plt.title("COVID-19")
plt.xlabel("Privacy Budget")
plt.ylabel("Acc. in %")
plt.legend()
plt.show()
```
